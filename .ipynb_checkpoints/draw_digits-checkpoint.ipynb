```python
import numpy as np
from pandas import read_csv
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

rawtrain = read_csv('train.csv', nrows = 10000).values

pca = PCA(n_components=196)
pca.fit(rawtrain[:, 1:])
```

```python
n = np.random.randint(0, rawtrain.shape[0])

sample = rawtrain[n, 1:]
sample = np.reshape(sample, (28, 28))

%matplotlib inline
plot = plt.imshow(sample)
plot.set_cmap('gray_r')
plt.show()
```

```python
sample = rawtrain[n, 1:]

sample = pca.transform(sample)

sample = np.reshape(sample, (14, 14))

%matplotlib inline
plot = plt.imshow(sample)
plot.set_cmap('gray_r')
plt.show()
```

```python
# testing progress bar

import pyprind
import time

start = time.time()
for j in range(3):
    print('Epoch ' + str(j + 1))
    for i in pyprind.prog_bar(range(100), stream=1):
        time.sleep(0.01)

overalltime = time.time() - start
overallminutes = str(overalltime // 60) + ' min, '
overallseconds = str(round(overalltime % 60, 3)) + ' sec'
print('Overall time:', overallminutes + overallseconds)
```
